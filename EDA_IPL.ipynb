# # Exploratory Data Analysis on Sports

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("matches.csv")
df.head()
```

```python
df=df[df['result']=='normal']
df.head()
```

```python
df['team1'].unique()
```

```python
df=df.replace('Delhi Daredevils','Delhi Capitals')
df['team1'].unique()
```

```python
df=df.replace('Rising Pune Supergiant','Rising Pune Supergiants')
df['team1'].unique()
```

```python
team1_matches_count=df.groupby('team1')['team1'].count()
team1_matches_count
```

```python
team2_matches_count=df.groupby('team2')['team2'].count()
team2_matches_count
```

```python
matches_per_team = team1_matches_count+team2_matches_count
matches_per_team
```

```python
matches_season = df.groupby('season')['season'].count()
matches_season
```

```python
seasons = df['season'].unique()
seasons
```

```python
seasons = np.sort(seasons)
seasons
```

```python
teams = df['team1'].unique()
teams
```

```python
df_toss_winner = df[df['toss_winner']==df['winner']]
df_toss_winner
```

```python
df_toss_match_winner_season = df_toss_winner.groupby('season')['season'].count() 
df_toss_match_winner_season
```

```python
df_toss_match_winner_team = df_toss_winner.groupby('winner')['winner'].count()
df_toss_match_winner_team
```

# Percentage matches won after winning toss

```python
toss_win_percent = []
for i in seasons:
    toss_win_percent.append(round((df_toss_match_winner_season[i]/matches_season[i])*100))
    
```

```python
toss_win_percent
```

```python
df_toss_win_percent_seasons=pd.DataFrame({'Seasons':seasons,'percentage':toss_win_percent})
df_toss_win_percent_seasons
```

```python
plt.figure(figsize=(12,6))
plt.bar(df_toss_win_percent_seasons['Seasons'],df_toss_win_percent_seasons['percentage'],color ='orange')
plt.title('Percentage matches won after winning toss')
plt.xlabel('Seasons')
plt.ylabel('Percentage')
plt.grid(True)
plt.show()
```

# Percentage matches won by teams after winning toss

```python
toss_win_team_percent=[]
for i in teams:
    toss_win_team_percent.append(round(df_toss_match_winner_team[i]/matches_per_team[i]*100))
```

```python
toss_win_team_percent
```

```python
df_toss_win_percent_teams = pd.DataFrame({'Teams':teams,'Percentage':toss_win_team_percent})
df_toss_win_percent_teams
```

```python
plt.tight_layout()
plt.figure(figsize=(12,6))
plt.barh(df_toss_win_percent_teams['Teams'],df_toss_win_percent_teams['Percentage'],color ='red')
plt.title('Percentage matches won by teams after winning toss')
plt.xlabel('Percentage')
plt.ylabel('Teams')
plt.grid(True)
plt.show()
```

# Wins toss,bats first & wins match by season

```python
df.head()
```

```python
tosswin_bat_matchwin = df[(df['toss_winner']==df['winner']) & (df['toss_decision']=='bat')]
tosswin_bat_matchwin
```

```python
tosswin_bat_matchwin_season = tosswin_bat_matchwin.groupby('season')['season'].count()
tosswin_bat_matchwin_season
```

```python
tosswin_bat_matchwin_winner = tosswin_bat_matchwin.groupby('winner')['winner'].count()
tosswin_bat_matchwin_winner
```

```python
tosswin_bat_winpercent_season = []
for i in seasons:
    tosswin_bat_winpercent_season.append(round((tosswin_bat_matchwin_season[i]/df_toss_match_winner_season[i])*100))
```

```python
tosswin_bat_winpercent_season
```

```python
df_tosswin_bat_teamwin_percent_seasons=pd.DataFrame({'Seasons':seasons,'percentage':tosswin_bat_winpercent_season})
df_tosswin_bat_teamwin_percent_seasons
```

```python
plt.figure(figsize=(12,6))
plt.bar(df_tosswin_bat_teamwin_percent_seasons['Seasons'],df_tosswin_bat_teamwin_percent_seasons['percentage'],color ='orange')
plt.title('Percentage Wins toss,bats first & wins match by season')
plt.xlabel('Seasons')
plt.ylabel('Percentage')
plt.grid(True)
plt.show()
```

# Percentage Wins toss,bats first & wins match by Teams

```python
tosswin_bat_matchwin_winner = tosswin_bat_matchwin.groupby('winner')['winner'].count()
tosswin_bat_matchwin_winner
```

```python
tosswin_bat_winpercent_team = []
for i in teams:
    if i in tosswin_bat_matchwin_winner.keys() and i in df_toss_match_winner_team.keys():
        tosswin_bat_winpercent_team.append(round(tosswin_bat_matchwin_winner[i]/df_toss_match_winner_team[i]*100))
    else:
        tosswin_bat_winpercent_team.append(0)
            
```

```python
tosswin_bat_winpercent_team
```

```python
teams
```

```python
toss_bat_winning = pd.DataFrame({'Teams':teams,'Percentage':tosswin_bat_winpercent_team})
toss_bat_winning
```

```python
plt.tight_layout()
plt.figure(figsize=(12,6))
plt.barh(toss_bat_winning['Teams'],toss_bat_winning['Percentage'],color ='red')
plt.title('Percentage of batting first while winning both toss and match')
plt.xlabel('Percentage')
plt.ylabel('Teams')
plt.grid(True)
plt.show()
```
